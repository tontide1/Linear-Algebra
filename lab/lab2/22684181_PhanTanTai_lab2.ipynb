TÃ­nh luá»¹ thá»«a báº­c k cá»§a ma tráº­n $ğ‘­_k = \left( \begin{matrix} 1&1 \\ 1&0 \end{matrix}\right)^\text{k}$ vÃ  chá»©ng minh ma trÃ¢nh **F_k** chá»©a 3 pháº§n tá»­ cá»§a $\text{ğ¹ğ‘–ğ‘ğ‘œ}_{ğ‘˜+1}$, $\text{ğ¹ğ‘–ğ‘ğ‘œ}_ğ‘˜$ vÃ  $\text{ğ¹ğ‘–ğ‘ğ‘œ}_{ğ‘˜âˆ’1}$ cá»§a dÃ£y Fibonacci (báº¯t Ä‘áº§u báº±ng 0 tá»« pháº§n tá»­ 0)

```python
import numpy as np
def fibonacci_matrix_power(k):
    F = np.array([[1, 1], [1, 0]])
    F_k = np.linalg.matrix_power(F, k)
    return F_k

def fibonacci_sequence_elements(k):
    F_k = fibonacci_matrix_power(k)
    F_iplus1 = F_k[0, 0]
    F_i = F_k[1, 0]
    F_iminus1 = F_k[1, 1]
    return F_iplus1, F_i, F_iminus1

k = int (input("nháº­p pháº§n tá»­ thá»© k: "))
F_iplus1, F_i, F_iminus1 = fibonacci_sequence_elements(k)

print(f"F_{k+1} =", F_iplus1)
print(f"F_{k} =", F_i)
print(f"F_{k-1} =", F_iminus1)
```
